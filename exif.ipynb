## EXIF Metadata in Images

```python
import os
import cv2
import exif
import numpy as np
import matplotlib.pyplot as plt
```

### Function to check EXIF contents

```python
def check_exif(img):
    if img.has_exif:
        print('The image contains EXIF information.')
    else:
        print('EXIF info not found.')
```

### Load Image

```python
im_org = exif.Image('original.JPG')
check_exif(im_org)
```

```python
dir(im_org)
```

```python
emoji = cv2.imread('exif-data-image-thumbnail-not-updating.png')

encoded = cv2.imencode('.png', emoji)[1]

data_encode = np.array(encoded)
  
# Converting the array to bytes.
byte_encode = data_encode.tobytes()
```

```python
byte_encode
```

```python
im_org.get_thumbnail()
```

```python
thumbnail = im_org.get_thumbnail()
decoded = cv2.imdecode(np.frombuffer(thumbnail, np.uint8), -1)
print(type(decoded))
plt.imshow(decoded[...,::-1]);
```

```python
im_org.get_thumbnail = byte_encode
```
